```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
cases_df = pd.read_parquet("cases_final_v1.parquet.gzip", engine="fastparquet")
```

```python
cases_df.columns
```

```python
StateCount = pd.value_counts(cases_df['State'].values, sort=True) 
```

```python
plt.figure(figsize=(16, 8))
# StateCount.plot.barh()
sns.barplot(StateCount, StateCount.index)
plt.title("Number of cases by State", fontsize=20)
plt.xlabel("Number of Cases", fontsize=15)
plt.ylabel("State", fontsize=15)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.show()
```

```python
CasesByYear = cases_df['decision_date'].groupby(cases_df.decision_date.dt.year).agg(["count"])
```

```python
CasesByYear = CasesByYear.sort_index(ascending=True)
plt.figure(figsize=(20, 12))
plt.scatter( CasesByYear.index, CasesByYear['count'])
plt.xlabel("Year", fontsize=15)
plt.ylabel("Number of Cases", fontsize=15)
plt.title("Number of Cases per Year", fontsize=25)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.show()
```
