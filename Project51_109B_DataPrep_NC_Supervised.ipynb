```python
from pandas.io.json import json_normalize

def expand_json_columns(df, col):
    df_temp = json_normalize(df[col])
    df_temp.columns = list(map(lambda x : "{0}_{1}".format(col, x) , df_temp.columns))
    df_temp.index = df.index
    df_final = pd.concat([df, df_temp], axis=1)
    return df_final
    
# # Expanding Court Data to columns
# court_df = pd.json_normalize(cases_df['court'])
# court_df.columns = list(map(lambda x : "court_" + x , court_df.columns))
# court_df.index = cases_df.index
# # court_df
# cases_df_final = pd.concat([cases_df, court_df], axis=1)

# df_final = pd.concat([df_final, court_df], axis=1)

# df_final.head()
```

```python
def extract_opinion_type(x):
    bs = BeautifulSoup(x['data'], "html.parser")
    if bs and bs.opinion:
        return  { "type": bs.opinion['type'], 
                 "text": bs.opinion.text }
    return None    
```

```python
def extract_opinion(x):
    bs = BeautifulSoup(x['data'], "html.parser")
    if bs and bs.opinion:
        return bs.opinion.text
    return None
```

```python
cases_df_final['opinion'] = cases_df_final.casebody.apply(extract_opinion_type)
```

```python
def extract_headnotes(x):
    bs = BeautifulSoup(x['data'], "html.parser")
    headnotes = bs.find_all('headnotes')
    result = []
    if headnotes:
        for note in headnotes:
            result.append(note.text)
        return " ".join(result)
    return None
```

```python
extract_headnotes(cases_df_final.casebody[11273534])
```

```python
cases_df_final['headnotes'] = cases_df_final.casebody.apply(extract_headnotes)
```
